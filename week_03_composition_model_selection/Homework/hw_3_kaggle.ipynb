```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.metrics import roc_auc_score
```

```python
std_scaler = StandardScaler()


X = data.drop(columns=target_col).values
y = data[target_col].values

X = std_scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.2, random_state=4)
```

```python
# LogisticRegressionCV

log_reg_cv = LogisticRegressionCV(solver='lbfgs', cv=7, Cs=10, scoring='roc_auc', refit=True)

log_reg_cv.fit(X_train, y_train)

y_pred = log_reg_cv.predict_proba(X_test)[:, 1]
print(f'Logistic Regression CV ROC-AUC score on test data: {roc_auc_score(y_test, y_pred)}')
```

```python
# LogisticRegression

log_reg = LogisticRegression(C=0.7)

log_reg.fit(X_train, y_train)

y_pred = log_reg.predict_proba(X_test)[:, 1]
print(f'Logistic Regression ROC-AUC score on test data: {roc_auc_score(y_test, y_pred)}')
```

```python
import catboost


#    
data = pd.read_csv('./train.csv')

cat_cols_to_delete = ['Sex', 'HasMultiplePhoneNumbers', 'HasPhoneService', 'HasOnlineTV', 'HasDeviceProtection', 'HasOnlineBackup', 'HasMovieSubscription']
cat_cols_new = [feature for feature in cat_cols if feature not in cat_cols_to_delete]
cat_cols_new.remove('IsSeniorCitizen')
cat_cols_new_indexes = np.array([4, 5, 6, 7, 8, 9, 10, 11])

data = data.drop(columns=cat_cols_to_delete)

X = data.drop(columns=target_col).values
y = data[target_col].values
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.2, random_state=4)

cat_boost = catboost.CatBoostClassifier(n_estimators=1000, cat_features=cat_cols_new_indexes)
cat_boost.fit(X_train, y_train)

y_pred = cat_boost.predict_proba(X_test)[:, 1]
print(f'CatBoostClassifier ROC-AUC score on test data: {roc_auc_score(y_test, y_pred)}')
```
